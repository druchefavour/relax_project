#### Missing Values
Fill in the missing values in df_U1UE['visited'] with zeros

```python
df_U1UE['visited'].fillna(0, inplace=True)
df_U1UE.head()
```

Fill in the missing values in df_U1UE['time_stamp'] with values from another df_U1UE['creation_time']

```python
df_U1UE.loc[df_U1UE['time_stamp'].isnull(),'time_stamp'] = df_U1UE['creation_time']
df_U1UE.head()
```

* Create a column in df_U1UE (Days_from_creation_to_visit) to display the number of days from the first visit to the next visit (s) of the user. 

```python
df_U1UE["Days_from_creation_to_visit"] = (pd.to_datetime(df_U1UE['time_stamp'])-pd.to_datetime(df_U1UE['creation_time'])).dt.days
df_U1UE.head()
```

#### HOW MANY VISITS IN EACH 7 DAY PERIOD?
* Differentiate between users. Assign 1 if next row is the same user and 0 if not. Call the column user_tag

```python
df_U1UE['user-tag'] = np.where(df_U1UE['user_id'].shift(-1)>df_U1UE['user_id'],0,1)
df_U1UE.head()
```

* Create a column, 'Days-From-previous-visit' derived by subtracting the values of two adjacent rows for each user 

```python
df_U1UE['Days-From-previous-visit'] = np.where(((df_U1UE['user-tag']>0) & (df_U1UE['Days_from_creation_to_visit']>0)) , (df_U1UE['Days_from_creation_to_visit'] - df_U1UE['Days_from_creation_to_visit'].shift(+1)),
                                      df_U1UE['Days_from_creation_to_visit'])
df_U1UE.head()
```

* If df_U1UE['Days-From-previous-visit] is less than or equal to 7, then the visit was made within one day period.
* To determine the number of visits within single 7 day periods let us assign one to df_U1UE['Days-From-previous-visit] less than or equal to 7 and zero to the others.
* Call the new column '7-Day-Period-Visit'

```python
df_U1UE['7-Day-Period-Visit'] = np.where(((df_U1UE['Days-From-previous-visit'] > 0) & (df_U1UE['Days-From-previous-visit'] <=7)), 1, 0)
df_U1UE.head()
```

#### HOW MANY VISITED 3 CONSECUTIVE DAYS IN EACH 7 DAY PERIOD?


* Establish where the 2 days consecutive visits happened within 7 days.

```python
df_U1UE['2-cum-days-visit'] =np.where(df_U1UE['user-tag']>0, (df_U1UE['Days-From-previous-visit'] + df_U1UE['Days-From-previous-visit'].shift(+1)), 0)
df_U1UE.head()
```

* Establish where there are 3-days consecutive visits.

```python
df_U1UE['3-cum-days-visit'] = np.where(df_U1UE['user-tag']>0, (df_U1UE['2-cum-days-visit'] + df_U1UE['2-cum-days-visit'].shift(+1)), 0)
df_U1UE.head()
```

* Establish adopted users visit by assigning 1 to each 3 days consecutive visits within 7 days period 

```python
df_U1UE['adopted-user-visit'] = np.where((df_U1UE['3-cum-days-visit'] >0) & (df_U1UE['3-cum-days-visit'] <=7), 1 , 0)
df_U1UE.head()
```

```python
from matplotlib.colors import ListedColormap
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
cm = plt.cm.RdBu
cm_bright = ListedColormap(['#FF0000', '#0000FF'])

def points_plot(ax, Xtr, Xte, ytr, yte, clf, mesh=True, colorscale=cmap_light, cdiscrete=cmap_bold, alpha=0.1, psize=10, zfunc=False, predicted=False):
    h = .02
    X=np.concatenate((Xtr, Xte))
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))

    #plt.figure(figsize=(10,6))
    if zfunc:
        p0 = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 0]
        p1 = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
        Z=zfunc(p0, p1)
    else:
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    ZZ = Z.reshape(xx.shape)
    if mesh:
        plt.pcolormesh(xx, yy, ZZ, cmap=cmap_light, alpha=alpha, axes=ax)
    if predicted:
        showtr = clf.predict(Xtr)
        showte = clf.predict(Xte)
    else:
        showtr = ytr
        showte = yte
    ax.scatter(Xtr[:, 0], Xtr[:, 1], c=showtr-1, cmap=cmap_bold, s=psize, alpha=alpha,edgecolor="k")
    # and testing points
    ax.scatter(Xte[:, 0], Xte[:, 1], c=showte-1, cmap=cmap_bold, alpha=alpha, marker="s", s=psize+10)
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    return ax,xx,yy
```

### 8.0 Tuning the Model
For tuning the parameters of our model, we will use a mix of cross-validation and grid search. In Logistic Regression, the most important parameter to tune is the regularization parameter C. We will note that the regularization parameter is not always part of the logistic regression model.

For tuning the parameters of our model, we will use a mix of cross-validation and grid search. In Logistic Regression, the most important parameter to tune is the regularization parameter C. We will note that the regularization parameter is not always part of the logistic regression model.

We will now implement some code to perform model tuning and selecting the regularization parameter C.

We will use the following cv_score function to perform K-fold cross-validation and apply a scoring function to each test fold. In this incarnation we use accuracy score as the default scoring function.

```python
from sklearn.cross_validation import KFold
from sklearn.metrics import accuracy_score

def cv_score(clf, x, y, score_func=accuracy_score):
    result = 0
    nfold = 5
    for train, test in KFold(y.size, nfold): # split data into train/test groups, 5 times
        clf.fit(x[train], y[train]) # fit
        result += score_func(clf.predict(x[test]), y[test]) # evaluate score function on held-out data
    return result / nfold # average
```

#### 8.0.1 Kfold (Cross Validation)
In order to find a good model using all the key independent variables, we will use a given list of possible values of C. For each C, we will create a logistic regression model with that value of C. We will find the average score for this model using the cv_score function only on the training set (Xlr, ylr) and we will then pick the C with the highest average score

The goal is to find the best model parameters based only on the training set, without showing the model test set at all.

```python
from sklearn.cross_validation import train_test_split
Xlr, Xtestlr, ylr, ytestlr = train_test_split(df_master[['user_id','org_id']].values, (df_master['adopted-user'] == 1).values)
```

```python
#the grid of parameters to search over
Cs = [0.001, 0.1, 1, 10, 100]
from sklearn.linear_model import LogisticRegression
max_score = 0

for C in Cs:
        clf = LogisticRegression(C=C)
        score = cv_score(clf, Xlr, ylr)

        if score > max_score:
            max_score = score
            best_C =C
print (max_score, best_C)
```

Here we see that the maximum score is 0.82 while the best C is 0.001. 
We will now evaluate the accuracy score of our prediction.

```python
clfl=LogisticRegression(C=best_C)
clfl.fit(Xlr, ylr)
ypred=clfl.predict(Xtestlr)
accuracy_score(ypred, ytestlr)
```

#### 8.0.2 Grid Search CV
Using grid search CV, we have:

```python
from sklearn.grid_search import GridSearchCV
clfl2=LogisticRegression()
parameters = {"C": [0.0001, 0.001, 0.1, 1, 10, 100]}
fitmodel = GridSearchCV(clfl2, param_grid=parameters, cv=5, scoring="accuracy")
fitmodel.fit(Xlr, ylr)
fitmodel.best_estimator_, fitmodel.best_params_, fitmodel.best_score_, fitmodel.grid_scores_
```

We observe that the best C is 0.0001 and the maximum score is 0.82. Now we generate the accuracy score of our prediction

```python
clfl2=LogisticRegression(C=fitmodel.best_params_['C'])
clfl2.fit(Xlr, ylr)
ypred2=clfl2.predict(Xtestlr)
accuracy_score(ypred2, ytestlr)
```

In these cases, we have used a squared error loss function along with Empirical Risk Minimization (ERM) to carry out regression. The idea there was to calculate this risk on the training set and minimize it. Then the hope was that on the population, or any testing set representative of it, the out-of-sample risk was similar in size to the in-sample training risk, and thus small. The question then is what might be an appropriate risk for classification? One immediately comes to mind: the fraction of misclassified samples.
